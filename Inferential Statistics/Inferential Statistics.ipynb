```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv('../Data/cleaned_data.csv')
df = df.drop(['Unnamed: 0'], axis=1)
df = df.dropna(subset = ['PRICE IN K', 'BEDS', 'BATHS', 'SQUARE FEET', 'LOT SIZE', '$/SQUARE FEET'])
```

```python
San_mateo = df.loc[df['COUNTY'] == 'SAN MATEO']
Santa_clara = df.loc[df['COUNTY'] == 'SANTA CLARA']
Alameda = df.loc[df['COUNTY'] == 'ALAMEDA']
San_Francisco = df.loc[df['COUNTY'] == 'SAN FRANCISCO']
```

```python
San_mateo_mean = San_mateo['PRICE IN K'].mean()
Santa_clara_mean = Santa_clara['PRICE IN K'].mean()
Alameda_mean = Alameda['PRICE IN K'].mean()
San_Francisco_mean = San_Francisco['PRICE IN K'].mean()
print('San_mateo_mean: ', San_mateo_mean)
print('Santa_clara_mean: ', Santa_clara_mean)
print('Alameda_mean: ', Alameda_mean)
print('San_Francisco_mean: ', San_Francisco_mean)
```

** House Price in Different Counties **

```python
stats.f_oneway(San_mateo['PRICE IN K'], Alameda['PRICE IN K'], Santa_clara['PRICE IN K'], San_Francisco['PRICE IN K'])
```

```python
# statistical significance
print(stats.f_oneway(San_mateo['PRICE IN K'], Alameda['PRICE IN K']))
print(stats.f_oneway(San_mateo['PRICE IN K'], Santa_clara['PRICE IN K']))
print(stats.f_oneway(San_mateo['PRICE IN K'], San_Francisco['PRICE IN K']))

print(stats.f_oneway(Santa_clara['PRICE IN K'], Alameda['PRICE IN K']))
print(stats.f_oneway(Santa_clara['PRICE IN K'], San_Francisco['PRICE IN K']))

print(stats.f_oneway(Alameda['PRICE IN K'], San_Francisco['PRICE IN K']))
```

** Property Type & Price per Square Foot **

```python
Single_family = df.loc[df['PROPERTY TYPE'] == 'SFH']
Condo = df.loc[df['PROPERTY TYPE'] == 'Condo']
Townhouse = df.loc[df['PROPERTY TYPE'] == 'TH']
```

```python
Single_family_mean = Single_family['$/SQUARE FEET'].mean()
Condo_mean = Condo['$/SQUARE FEET'].mean()
Townhouse_mean = Townhouse['$/SQUARE FEET'].mean()
print('Single family unit price: ', Single_family_mean)
print('Condo unit price: ', Condo_mean)
print('Townhouse unit price: ', Townhouse_mean)
```

```python
stats.f_oneway(Single_family['$/SQUARE FEET'], Condo['$/SQUARE FEET'], Townhouse['$/SQUARE FEET'])
```

```python
print(stats.f_oneway(Condo['$/SQUARE FEET'], Townhouse['$/SQUARE FEET']))
print(stats.f_oneway(Single_family['$/SQUARE FEET'], Condo['$/SQUARE FEET']))
print(stats.f_oneway(Single_family['$/SQUARE FEET'], Townhouse['$/SQUARE FEET']))
```

** Single Family House: Number of Bedrooms & House Price **

```python
# delete some outliers
Single_family = Single_family.sort_values(by='BEDS', ascending=False)

Single_family.reset_index(drop=True)
Single_family = Single_family.iloc[1:]

Single_family = Single_family.sort_values(by='BEDS')

Single_family.reset_index(drop=True)
Single_family = Single_family.iloc[8:]
```

```python
x = [a for a in Single_family['BEDS']]
y = list(Single_family['PRICE IN K'])
slope, intercept = np.polyfit(x, y, deg=1)
r = np.corrcoef(x, y)[0,1]
print('Slope:' , slope)
print('Intercept: ', intercept)
print('Pearson r: ', r)
```

```python
y_pred = [i * slope + intercept for i in x]
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2, color='green')
plt.plot(x, y_pred, linewidth=3, color='red')
ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of Bedrooms', fontsize=12)
ax.set_ylabel('House Price in K', fontsize=12)
ax.set_title('Single Family House: Number of Bedrooms vs. House Price', fontsize=14)

ax.grid(True)
fig.tight_layout()
plt.show()
```
